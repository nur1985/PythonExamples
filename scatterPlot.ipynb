```python
# Author: Nuruzzaman
# Date:   02/08/2019
# Scatter plot example with different options 

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
s = np.random.rand(*x.shape) * 300 + 500

plt.scatter(x, y, s, c='b', alpha=0.7, marker='o', label="Luck")
plt.xlabel("My X", fontsize=15, color='red', labelpad=15, horizontalalignment='right')
plt.ylabel("My Y", fontsize=20, backgroundcolor='#FFFF00')

plt.legend("Test", 
          title="Legend Title", fontsize =44, 
#          loc="upper left",
          ncol=2, mode="expand", borderaxespad=0., shadow=True, 
          bbox_to_anchor=(0.01, 0.01, 0.24, 0.965), 
          fancybox=True, framealpha=1.0, borderpad=1, prop={'size':15})
plt.xticks(rotation=70)
plt.tick_params(axis='x', direction='in',pad=5)
plt.tick_params(axis='y', direction='in',pad=5)

ax = plt.axes()        
ax.yaxis.grid(color='red', alpha=0.1, linestyle='--', linewidth=1) # horizontal lines
ax.xaxis.grid(color='grey', alpha=0.5, linestyle='-.', linewidth=2) # vertical lines
#ax.set_xlabel("My X", fontsize=15, color='red', labelpad=15, xy=(0, 1), xytext=(-15,2), ha='left', va='top', xycoords='axes fraction', textcoords='offset points' )
#ax.xaxis.set_label_coords(1.05, -0.025)
ax.annotate('My X1', xy=(1.0, -0.17), xytext=(1.05, -0.025), ha='right', va='bottom', xycoords='axes fraction', textcoords='offset points',fontsize=15, color='red')
ax.annotate('My Y1', rotation=90,xy=(-0.13, 1.00), xytext=(1.05, -0.025), ha='left', va='top', xycoords='axes fraction', textcoords='offset points',fontsize=15, color='red')

plt.show()
```
